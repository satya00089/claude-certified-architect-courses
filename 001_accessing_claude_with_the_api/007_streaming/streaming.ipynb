#### How streaming works

1. Open a streaming request with `client.messages.stream(...)`
2. Iterate over `stream.text_stream` to receive incremental text chunks
3. After iteration, use `stream.get_final_message()` to obtain the assembled message object


```python
from anthropic import Anthropic
from dotenv import load_dotenv

load_dotenv()
client = Anthropic()

# Example (commented out to avoid accidental API calls)
messages = [
    {"role": "user", "content": "Write a 1-sentence description of a fake database."}
]
with client.messages.stream(
    model="claude-sonnet-4-0", max_tokens=500, messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")

# final = stream.get_final_message()
# print("Final assembled text:")
# print(final.content[0].text)
```
